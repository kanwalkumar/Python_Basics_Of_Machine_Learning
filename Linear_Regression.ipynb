```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import fetch_california_housing
fetch_california_housing_data = fetch_california_housing()
```

```python
df_x=pd.DataFrame(fetch_california_housing_data.data, columns=fetch_california_housing_data.feature_names)
df_y=pd.DataFrame(fetch_california_housing_data.target)
```

```python
df_x.describe()
```

```python
reg=linear_model.LinearRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x, df_y, test_size= 0.2, random_state=4 )
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
#y = mx+c ==> m weights
```

```python
price = reg.predict(x_test)
```

```python
price[4100]
```

```python
comparison = pd.DataFrame({"Actual": np.array(y_test).flatten(), "Predicted": np.array(price).flatten()})
print(comparison.head(20))
```
